```python

import pandas as pds

# Select wanted category here. All category names are in the "files" on the sidebar
data = pds.read_csv('/content/amazon-us-customer-reviews-dataset/amazon_reviews_us_Apparel_v1_00.tsv', sep='\t', on_bad_lines='skip')
```

```python
counts = data['product_title'].value_counts()
```

```python
# the top 30 items with most reviews, select one of these.
print(counts.head(30))
```

```python
from sklearn.utils import shuffle
data = shuffle(data)
```

```python
# select amout of reviews wanted
count = 200
allcount = count;

# input the selected title of item.
name = "Champion Men's Full-zip Eco Fleece Hoodie Jacket"
```

```python
res = [];
for index, item in data.iterrows():
  if (item['product_title'] == name):
      count -= 1
      res.append(item);
  if count == 0:
    break;
```

```python
for index, item in data.iterrows():
  if (item['product_title'] == name):
    print(item);
    # print(item['product_title'].replace(' ',''))
    break;

stars = 0;
for item in res:
  stars += int(item['star_rating']);

print(' ')
print(stars/allcount);

# item data and also actual rating average.
```

```python
import pprint
import json
```

```python
jsonList = [];
for item in res:
  obj = {
    "title": item['review_headline'],
    "body": item['review_body'],
    "date": item['review_date'],
    "rating": 0,
    "likes": 0,
    "dislikes": 0,
    "account": {
        "id": 5
        },
    "item": {
        "id": 1
    }
  }
  jsonList.append(obj);

for item in jsonList:
  item['body'] = item['body'].replace('\\\\"', "'").replace('<br />', '').replace('"', "'")
  item['title'] = item['title'].replace('\\\\"', "'").replace('<br />', '').replace('"', "'")

print(json.dumps(jsonList, indent=4));

# Prints JSON in to the ready to go format. Just copy into a file.
```
